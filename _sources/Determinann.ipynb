### 1. Jika terdapat Matriks Identitas maka Determinannya adalah 1 berapapun ordonya




```python
print("Matriks A :")
print(b)

det = np.linalg.det(b)
print(f"Determinan dari Matriks A : {int(det.round())}")
```

# Cofactor


### Tugas Kofaktor matrix ordo 5x5
Matrix yang akan saya buat adalah matrix dengan ordo 5x5

ialah sebagai berikut
$$
A=\begin{bmatrix}
1 &2&1&1&1\\
2&1&3&1&1\\
3&2&3&2&2 \\
4&1&1&4&3\\
3&5&1&2&2
\end{bmatrix}
$$
Matrix ini akan saya jadikan invers, hal pertama yang akan cari adalah determinannya terlebih dahulu

Rumus Kofaktor=
$$
C_{ij} = (-1)^{i+j} * detA
$$

rumus det (pick one row or column) =
$$
a_{ij} * c_{ij} + a_{ij} * c_{ij} + a_{ij} * c_{ij} + a_{nn} * c_{nn}$$

rumus invers matrix =
$${ \frac{1}{detA}} * adj(a)$$

Mencari Adjoin =
$$adj(A) = (kof(A))^T$$

(adj(S) = Transpose dari setiap cofator-cofator Matrix)
