Matriks invers adalah konsep dalam aljabar linear yang terkait dengan matriks persegi (matriks yang memiliki jumlah baris dan kolom yang sama). Matriks invers dari suatu matriks A, jika ada, dilambangkan dengan A^(-1).

Jika A adalah sebuah matriks persegi n x n, maka matriks invers A^(-1) memiliki sifat bahwa ketika A dikalikan dengan A^(-1), hasilnya adalah matriks identitas, disimbolkan sebagai I:

A * A^(-1) = A^(-1) * A = I

\a
Matriks identitas (I) adalah matriks persegi dengan semua elemen diagonalnya adalah 1, dan elemen di luar diagonal adalah 0.

Matriks invers hanya dapat ditemukan untuk matriks yang non-singular, yang berarti determinannya tidak nol. Dalam kasus matriks singular (determinannya nol), matriks tidak memiliki invers.

Matriks invers memungkinkan untuk membalikkan efek transformasi yang dilakukan oleh matriks asli. Ini berguna dalam berbagai aplikasi, termasuk sistem persamaan linear, pemrosesan sinyal, dan pemodelan matematis, di mana seringkali kita perlu "membalikkan" efek suatu transformasi yang telah diterapkan pada data.
